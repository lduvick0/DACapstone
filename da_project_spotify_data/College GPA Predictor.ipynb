```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Step 1: Load the data
data = pd.read_csv("C:/Users/micha/PycharmProjects/Docs/sat_gpa_v1.csv")
data.head()
```

```python
# The height and aspect parameters control the size and aspect ratio of the plot, respectively
# The scatter_kws={'alpha':0.5} parameter adds opacity to the markers, making them semi-transparent.
sns.lmplot(x='sat_sum', y='fy_gpa', data=data, height=6, aspect=1.5,scatter_kws={'alpha':0.5})
plt.title('Relationship between High School SAT Scores and College GPA')
plt.xlabel('SAT Scores')
plt.ylabel('GPA')

# Linear regression model
# It creates a linear regression model using sm.OLS() from statsmodels (sm).
# Adding a constant term allows the regression line to have a y-intercept, meaning that the line does not necessarily have to pass through the origin 

X = sm.add_constant(data['sat_sum'])
y = data['fy_gpa']
# sm.OLS(y, X): This part of the code creates an Ordinary Least Squares (OLS) regression model using the sm (statsmodels) library.
# .fit(): This method fits the OLS model to the data. It estimates the coefficients of the regression equation using the Ordinary Least Squares method
model = sm.OLS(y, X).fit()
plt.show()
# prints a summary of the linear regression model using model.summary(). 
print(model.summary())
```
